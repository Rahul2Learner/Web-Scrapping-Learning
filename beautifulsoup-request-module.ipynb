```python
from bs4 import BeautifulSoup
import requests
```

```python
URL="https://quotes.toscrape.com/"
```

```python
response= requests.get(url = URL)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.text,parser='html.parser')
```

```python
all_quotes = soup.find_all(class_="quote")
```

```python
len(all_quotes)
```

```python
first_quote = all_quotes[0]
```

```python
first_quote
```

```python
quote = first_quote.find("span", class_="text").text
quote
```

```python
author = first_quote.find("small",class_= 'author',attrs={'itemprop':'author'}).text
author
```

```python
list_all_tags= [i.text for i in first_quote.find_all("a",class_= 'tag')]
list_all_tags
```

```python

scrapped_data = []
for quote in all_quotes:
    q = quote.find("span", class_="text").text
    author = quote.find("small",class_= 'author',attrs={'itemprop':'author'}).text
    list_all_tags= [i.text for i in quote.find_all("a",class_= 'tag')]

    scrapped_data.append((q,author,list_all_tags))
    
    
```

```python
for i in scrapped_data:
    print(i)
```
